````python
def initialize_inventory(products):
    inventory = {}
    for product in products:
        valid_quantity = False
        while not valid_quantity:
            try:
                quantity = int(input(f"Enter the quantity of {product}s available: "))
                if quantity < 0:
                    raise ValueError("Invalid quantity! Please enter a non-negative value.")
                valid_quantity = True
            except ValueError as error:
                print(f"Error: {error}")
        inventory[product] = quantity
    return inventory

def calculate_total_price(prices):
    total_price = 0
    for product, price in prices.items():
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price of {product}: "))
                if price < 0:
                    raise ValueError("Invalid price! Please enter a non-negative value.")
                prices[product] = price
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}")
    
    return total_price

def get_customer_orders(inventory):
    orders = {}
    while True:
        valid_order = False
        while not valid_order:
            try:
                product = input("Enter the product name: ").strip()
                if product not in inventory:
                    raise ValueError(f"Product '{product}' does not exist in inventory.")
                quantity = int(input(f"Enter the quantity of {product} to order: "))
                if quantity <= 0:
                    raise ValueError("Quantity must be greater than zero.")
                if quantity > inventory[product]:
                    raise ValueError(f"Insufficient stock for {product}. Available quantity is {inventory[product]}.")
                orders[product] = quantity
                inventory[product] -= quantity
                valid_order = True
            except ValueError as error:
                print(f"Error: {error}")
        
        more_orders = input("Do you want to add more orders? (yes/no): ").strip().lower()
        if more_orders != 'yes':
            break
    
    return orders

# Example usage:
products = ["t-shirt", "mug", "hat", "book", "keychain"]
inventory = initialize_inventory(products)

# Assume prices are initially set to 0 and will be updated later
prices = {product: 0 for product in products}
calculate_total_price(prices)

orders = get_customer_orders(inventory)
print(f"Orders: {orders}")
print(f"Remaining Inventory: {inventory}")
````
